## Get filings

## How to get filings:
1. Declare a variable to get filings as .get_filings()
2. Filter the desired filings as .filter(form = ' ', date = ' ')
3. Make that filing into a obj = .obj()
4. Access financial documents using .financials
5. There are 3 types of financial documents 
- company's balance sheet = financials.balance_sheet()
- company's income statement = financials.income_statement()
- company's cash flow statement = financials.cash_flow_statement()
